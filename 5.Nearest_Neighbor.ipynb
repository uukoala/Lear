# Nearest Neighbor

```python
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
#import data
mnist = input_data.read_data_sets("MNIST/", one_hot=True)
```

```python
Xtr, Ytr = mnist.train.next_batch(5000)
Xte, Yte = mnist.test.next_batch(200)

xtr = tf.placeholder(tf.float32, [None, 784])
xte = tf.placeholder(tf.float32, [784])
#Count the distance between test and every train example
distance = tf.reduce_sum(tf.abs(xte -xtr), reduction_indices = 1)
#Find the nearest one index
pred = tf.argmin(distance, 0)
```

```python
init = tf.initialize_all_variables()
with tf.Session() as sess:
    sess.run(init)
    #Count the numbers of true prediction
    count = 0.0
    for i in range(len(Xte)):
        nn_index = sess.run(pred, feed_dict={xtr: Xtr, xte: Xte[i]})
        print "Test: %d, Prediction: %d, True Classify: %d" \
                    %(i+1, np.argmax(Ytr[nn_index],0), np.argmax(Yte[i],0))
        if np.argmax(Ytr[nn_index],0) == np.argmax(Yte[i], 0):
            count += 1
    print "Done"
    print "Accuracy:", count/len(Xte)
        
```
